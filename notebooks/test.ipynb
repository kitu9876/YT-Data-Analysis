```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage containing the YouTube channel link
url = "https://www.youtube.com/watch?v=cnY6jQlJcQc"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the YouTube channel link based on its attributes or text content
    # Replace 'attribute_name' and 'attribute_value' with appropriate values
    youtube_channel_link = soup.find('a', attrs={'class': 'yt-simple-endpoint style-scope yt-formatted-string'})
    
    if youtube_channel_link:
        # Extract the actual link from the anchor tag
        channel_link = youtube_channel_link['href']
        print("YouTube Channel Link:", channel_link)
    else:
        print("YouTube channel link not found on the page.")
else:
    print("Failed to retrieve the webpage.")
```

```python
import re

def extract_channel_id(video_url):
    # Regular expression pattern to match YouTube video URLs
    pattern = r'(?:https?://)?(?:www\.)?youtube\.com/watch\?v=([\w-]+)'
    
    # Search for the video ID using the pattern
    match = re.search(pattern, video_url)
    
    if match:
        video_id = match.group(1)
        
        # Construct the channel URL based on the video ID
        channel_url = f"https://www.googleapis.com/youtube/v3/videos?id={video_id}&key=YOUR_API_KEY&part=snippet"
        
        # Send a request to the API and retrieve the JSON response
        response = requests.get(channel_url)
        data = response.json()
        
        # Extract the channel ID from the JSON response
        if "items" in data and len(data["items"]) > 0:
            channel_id = data["items"][0]["snippet"]["channelId"]
            return channel_id
    
    return None

# Example YouTube video URL
video_url = "https://www.youtube.com/watch?v=VIDEO_ID_HERE"

# Extract and print the channel ID
channel_id = extract_channel_id(video_url)
if channel_id:
    print("Channel ID:", channel_id)
else:
    print("Channel ID not found.")
```
